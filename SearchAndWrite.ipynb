```python
import os
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool
```

```python
search_tool = SerperDevTool()
```

```python
from langchain_openai import ChatOpenAI
```

```python
# Define your agents with roles and goals
researcher = Agent(
    role = "Senior Research Assistant",
    goal = "Look up the latest Advancements in AI Agents",
    backstory = """ You work at a leading tech think tank.
    Your expertise lies in identifying emerging trends.
    you have a knack for dissecting complex data and presenting actionable insights.""",
    verbose = False,
    allow_delegation = False,
    tools = [search_tool],
    llm = ChatOpenAI(model_name = "gpt-4-turbo-preview", temperature = 0.3)

)
```

```python
writer = Agent(
    role = 'Professional Short-Article Writer',
    goal = 'Summarize the latest advancement int AI Agents in a concise article',
    backstory = """You are a renowned Content Strategist, known for your insightfil and engaging articles.
    You transform complex concepts into compelling narratives.""",
    verbose = True,
    allow_delegation = True,
    llm = ChatOpenAI(model_name = "gpt-4-turbo-preview", temperature = 0.7)

)
```

```python
#Create tasks for your agents
task1 = Task(
    description = """Condut a comprehensive analysis of the latest
    advance in AI Agents in April.
    Identify key trends breakthrough technologies, and potential industry impact.""",
    expected_output = "Full analysis repot in bullet points",
    agent = researcher

)
```

```python
task2 = Task(
    description = """Using the insights provided, write a short article that highlights the most significant AI Agent advancements.
    your post should be informative yet accessible, catering to a tech-savvy audience.
    Make it sound cool, avoid complex words so it doesn't sound like AI.""",
    expected_output = "Full blog post of at least 3 paragraphs",
    agent=writer

)
```

```python
#Instantiate your crew with a sequential process
crew = Crew(
    agents=[researcher,writer],
    tasks=[task1, task2],
    verbose = 1,# You can set it to 1 or 2 to different logging levels

)
```

```python
result = crew.kickoff()
print("######")
print(result)
```

```python
result = crew.kickoff()
print("######")
print(result)
```
